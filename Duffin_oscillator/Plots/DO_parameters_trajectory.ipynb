#  Duffing oscillator process in a ball of $\mathbb{R}^2$ : visualization of trajectories and Monte Carlo estimation of MET

Based on the paper [A neural network solution of first-passage problems](https://link.springer.com/article/10.1007/s10483-024-3189-8) (Jiamin Qian, Lincong Chen & J. Q. Sun, Oct. 2024), the 2-dimensional Duffing oscillator is defined by:
$$ d \begin{pmatrix} X_1(t)\\
X_2(t) \end{pmatrix} = \begin{pmatrix} X_2\\
-X_1 - X_1^3 - 2\zeta X_2 \end{pmatrix} dt + \begin{pmatrix} \sqrt{2\varepsilon}dB_1(t)\\
\sqrt{2\zeta} dB_2(t) \end{pmatrix}$$

## 1) Visualization of the process:

```python
#parameters of simulation
bm_0 = np.zeros(2)
time_horizon = 100
dt = 10**-3
n_steps = int(time_horizon/dt)

# Brownian motion in a ball
bm = np.zeros((2,n_steps))
bm[:, 0] = bm_0

for k in range(1, n_steps):
    bm[:, k] = bm[:, k-1] + np.random.normal(scale=np.sqrt(dt), size=2)

#sanity check
plt.plot(bm[0], bm[1])
```

```python
#Duffin-oscillator process

#parameters
do_0 = np.zeros(2)
zeta = 0.08
eps = 10**-5


do = np.zeros((2,n_steps))
do[:, 0] = do_0

for k in range(1, n_steps):
    scaled_b1 = np.sqrt(2*eps)*np.random.normal(scale=np.sqrt(dt))
    scaled_b2 = np.sqrt(2*zeta)*np.random.normal(scale=np.sqrt(dt))
    do[:, k] = do[:, k-1] \
              + dt*np.array([do[1,k-1], - do[0,k-1] - do[0,k-1]**3 - 2*zeta*do[1,k-1] ])\
              + np.array([scaled_b1, scaled_b2])

#sanity check
plt.plot(do[0], do[1])
```

```python
#truncating the realizations outside the circle
radius = 2
do_norms = np.zeros(n_steps)
bm_norms = np.zeros(n_steps)

for k in range(1,n_steps):
    do_norms[k] = np.linalg.norm(do[:,k])
    bm_norms[k] = np.linalg.norm(bm[:,k])

do_exit = np.argmax(do_norms >= radius)
bm_exit = np.argmax(bm_norms >= radius)

if do_exit == 0:
    do_exit = n_steps

if bm_exit == 0:
    bm_exit = n_steps

trunc_do = np.zeros((2,do_exit))
trunc_bm = np.zeros((2,bm_exit))

trunc_do[:,:] = do[:,0:do_exit]
trunc_bm[:,:] = bm[:,0:bm_exit]
```

```python
do_exit_time = do_exit * dt
bm_exit_time = bm_exit * dt
```

```python
#plotting the brownian motion and Dufflin Oscillator process together

fig, ax = plt.subplots(figsize=(5,5))
ax.plot(trunc_bm[0],trunc_bm[1], c='b', label = "Standard BM")
ax.plot(trunc_do[0], trunc_do[1], c='g', alpha=0.7, label = "DO process")
ax.plot(trunc_do[0,-1], trunc_do[1,-1], 'k+', mew=10, ms=5)
ax.plot(trunc_bm[0,-1], trunc_bm[1,-1], 'k+', mew=10, ms=5)
circle = plt.Circle((0,0),radius , fill = False)
ax.add_artist(circle)
ax.set_xlim([-1.1*radius, 1.1*radius])
ax.set_ylim([-1.1*radius, 1.1*radius])
ax.legend()
#plt.autoscale(False, tight=False)
plt.show()
```

```python
print(f"BM exit time: {bm_exit_time}s, DO exit time: {do_exit_time}s.")
```

## 2) Influence of the parameters on the MET

```python
def duffin_oscillator(zeta, eps, radius, do_0, time_horizon, dt, exit=True):
    #initialization
    n_steps = int(time_horizon/dt)
    do = np.zeros((2,n_steps))
    do[:, 0] = do_0

    #generating the whole path
    for k in range(1, n_steps):
        scaled_b1 = np.sqrt(2*eps)*np.random.normal(scale=np.sqrt(dt))
        scaled_b2 = np.sqrt(2*zeta)*np.random.normal(scale=np.sqrt(dt))
        do[:, k] = do[:, k-1] \
                  + dt*np.array([do[1,k-1], - do[0,k-1] - do[0,k-1]**3 - 2*zeta*do[1,k-1] ])\
                  + np.array([scaled_b1, scaled_b2])

    #truncating the trajectory outside the circle
    do_norms = np.zeros(n_steps)
    for k in range(1,n_steps):
        do_norms[k] = np.linalg.norm(do[:,k])

    do_exit = np.argmax(do_norms >= radius)
    if do_exit == 0:
        do_exit = n_steps
        exit = False

    trunc_do = np.zeros((2,do_exit))
    trunc_do[:,:] = do[:,0:do_exit]

    return exit, trunc_do
```

```python
#choose a range of zeta eps
zetas = [0.05 * 10**k for k in range(3)]
epsis = [1e-8 * 10**(2*k) for k in range(3)]
n, m = len(zetas), len(epsis)

#parameters of the simulation
do_0 = np.zeros(2)
time_horizon = 200
dt = 10**-3
radius = 2

#plot a circle
theta = np.linspace(0, 2*np.pi, 100)
xs = radius*np.cos(theta)
ys = radius*np.sin(theta)

#plot
fig, axs = plt.subplots(n, m, figsize=(15, 15))
#fig.set_figheight(13)
#fig.set_figwidth(11)
for i in range(n):
    for j in range(m):
        exit_ij, do_ij = duffin_oscillator(zetas[i], epsis[j], radius, do_0, time_horizon, dt, exit=True)

        if exit_ij == True:
            axs[i,j].plot(do_ij[0,-1], do_ij[1,-1], 'k+', mew=10, ms=5)
            exit_message = f"exit time: {len(do_ij[0,:])*dt:.3f}s"
        else:
            exit_message = f"exit time: >{time_horizon}s!"

        axs[i,j].plot(do_ij[0], do_ij[1])
        axs[i,j].set_title(f"zeta={zetas[i]}, epsilon={epsis[j]}\n" + exit_message)
        axs[i,j].plot(xs,ys)
        #axs[i,j].axis('equal')
        axs[i,j].set_xlim([-1.1*radius, 1.1*radius])
        axs[i,j].set_ylim([-1.1*radius, 1.1*radius])
```

```python
import numpy as np
import matplotlib.pyplot as plt

def duffin_oscillator(zeta, eps, radius, do_0, time_horizon, dt):
    """Simulate a Duffing oscillator with noise and truncate trajectory if it exits the ball."""
    n_steps = int(time_horizon / dt)
    do = np.zeros((2, n_steps))
    do[:, 0] = do_0

    # Generate the trajectory
    for k in range(1, n_steps):
        scaled_b1 = np.sqrt(2 * eps) * np.random.normal(scale=np.sqrt(dt))
        scaled_b2 = np.sqrt(2 * zeta) * np.random.normal(scale=np.sqrt(dt))
        do[:, k] = do[:, k - 1] + dt * np.array([
            do[1, k - 1],
            -do[0, k - 1] - do[0, k - 1]**3 - 2 * zeta * do[1, k - 1]
        ]) + np.array([scaled_b1, scaled_b2])

    # Check norms and truncate trajectory
    norms = np.linalg.norm(do, axis=0)
    exit_idx = np.argmax(norms >= radius)
    if exit_idx == 0:  # If no exit occurs, return full trajectory
        exit_idx = n_steps
        exited = False
    else:
        exited = True

    return exited, do[:, :exit_idx]

# Parameters
zetas = np.linspace(0.3, 3, 9).tolist()  # Range of zeta values
eps = 1e-3  # Fixed epsilon value
do_0 = np.zeros(2)  # Initial condition
time_horizon = 200
dt = 1e-3
radius = 2

# Circle for the boundary
theta = np.linspace(0, 2 * np.pi, 100)
xs = radius * np.cos(theta)
ys = radius * np.sin(theta)

# Plot setup
fig, axes = plt.subplots(len(zetas), 1, figsize=(10, 15), constrained_layout=True)
fig.suptitle("Duffing Oscillator Trajectories for Different Zeta Values", fontsize=16, weight="bold")

# Plot trajectories for each zeta
for i, zeta in enumerate(zetas):
    exited, trajectory = duffin_oscillator(zeta, eps, radius, do_0, time_horizon, dt)

    # Plot trajectory
    ax = axes[i]
    ax.plot(trajectory[0], trajectory[1], label="Duffing Oscillator", color="royalblue", alpha=0.8)
    ax.plot(xs, ys, 'k--', linewidth=1, alpha=0.7, label="Boundary")

    # Mark endpoint
    if exited:
        ax.plot(trajectory[0, -1], trajectory[1, -1], 'ro', label="Exit Point")
        exit_message = f"Exit time: {len(trajectory[0]) * dt:.2f}s"
    else:
        exit_message = f"Exit time: >{time_horizon}s!"

    # Title and labels
    ax.set_title(f"$\\zeta={zeta:.2f}$, $\\epsilon={eps}$\n{exit_message}", fontsize=10, weight="bold")
    ax.set_xlim([-1.1 * radius, 1.1 * radius])
    ax.set_ylim([-1.1 * radius, 1.1 * radius])
    ax.grid(color='lightgray', linestyle='--', linewidth=0.5, alpha=0.5)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.legend(loc="upper right", fontsize=8)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def duffin_oscillator(zeta, eps, radius, do_0, time_horizon, dt):
    """Simulate a Duffing oscillator with noise and truncate trajectory if it exits the ball."""
    n_steps = int(time_horizon / dt)
    do = np.zeros((2, n_steps))
    do[:, 0] = do_0

    # Generate the trajectory
    for k in range(1, n_steps):
        scaled_b1 = np.sqrt(2 * eps) * np.random.normal(scale=np.sqrt(dt))
        scaled_b2 = np.sqrt(2 * zeta) * np.random.normal(scale=np.sqrt(dt))
        do[:, k] = do[:, k - 1] + dt * np.array([
            do[1, k - 1],
            -do[0, k - 1] - do[0, k - 1]**3 - 2 * zeta * do[1, k - 1]
        ]) + np.array([scaled_b1, scaled_b2])

    # Check norms and truncate trajectory
    norms = np.linalg.norm(do, axis=0)
    exit_idx = np.argmax(norms >= radius)
    if exit_idx == 0:  # If no exit occurs, return full trajectory
        exit_idx = n_steps
        exited = False
    else:
        exited = True

    return exited, do[:, :exit_idx]

def brownian_motion(radius, bm_0, time_horizon, dt):
    """Simulate a standard Brownian motion and truncate trajectory if it exits the ball."""
    n_steps = int(time_horizon / dt)
    bm = np.zeros((2, n_steps))
    bm[:, 0] = bm_0

    for k in range(1, n_steps):
        bm[:, k] = bm[:, k - 1] + np.random.normal(scale=np.sqrt(dt), size=2)

    # Check norms and truncate trajectory
    norms = np.linalg.norm(bm, axis=0)
    exit_idx = np.argmax(norms >= radius)
    if exit_idx == 0:  # If no exit occurs, return full trajectory
        exit_idx = n_steps
        exited = False
    else:
        exited = True

    return exited, bm[:, :exit_idx]

# Parameters
zetas = np.linspace(0.3, 3, 9).tolist()  # Range of zeta values
eps = 1e-3  # Fixed epsilon value
do_0 = np.zeros(2)  # Initial condition for Duffing oscillator
bm_0 = np.zeros(2)  # Initial condition for Brownian motion
time_horizon = 200
dt = 1e-3
radius = 2

# Circle for the boundary
theta = np.linspace(0, 2 * np.pi, 100)
xs = radius * np.cos(theta)
ys = radius * np.sin(theta)

# Plot setup
fig, axes = plt.subplots(3, 3, figsize=(15, 15), constrained_layout=True)
fig.suptitle("Duffing Oscillator and Brownian Motion Trajectories", fontsize=16, weight="bold")

# Plot trajectories for each zeta
for idx, zeta in enumerate(zetas):
    i, j = divmod(idx, 3)
    ax = axes[i, j]

    # Duffing oscillator
    exited_do, trajectory_do = duffin_oscillator(zeta, eps, radius, do_0, time_horizon, dt)
    if exited_do:
        exit_time_do = f"{len(trajectory_do[0]) * dt:.2f}s"
    else:
        exit_time_do = f">{time_horizon}s"

    # Brownian motion
    exited_bm, trajectory_bm = brownian_motion(radius, bm_0, time_horizon, dt)
    if exited_bm:
        exit_time_bm = f"{len(trajectory_bm[0]) * dt:.2f}s"
    else:
        exit_time_bm = f">{time_horizon}s"

    # Plot Duffing oscillator trajectory
    ax.plot(trajectory_do[0], trajectory_do[1], label="Duffing Oscillator", color="royalblue", alpha=0.8)
    if exited_do:
        ax.plot(trajectory_do[0, -1], trajectory_do[1, -1], 'ro', label="Duffing Exit Point")

    # Plot Brownian motion trajectory
    ax.plot(trajectory_bm[0], trajectory_bm[1], label="Brownian Motion", color="seagreen", alpha=0.8)
    if exited_bm:
        ax.plot(trajectory_bm[0, -1], trajectory_bm[1, -1], 'go', label="Brownian Exit Point")

    # Circle boundary
    ax.plot(xs, ys, 'k--', linewidth=1, alpha=0.7, label="Boundary")

    # Title and labels
    ax.set_title(
        f"$\\zeta={zeta:.2f}$, $\\epsilon={eps}$\nDuffing Exit: {exit_time_do}, BM Exit: {exit_time_bm}",
        fontsize=10, weight="bold"
    )
    ax.set_xlim([-1.1 * radius, 1.1 * radius])
    ax.set_ylim([-1.1 * radius, 1.1 * radius])
    ax.grid(color='lightgray', linestyle='--', linewidth=0.5, alpha=0.5)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.legend(loc="upper right", fontsize=8)

plt.show()
```
